# 图像的剪切、粘贴与合并操作

```python
im = Image.open('demo.jpg')
box = (50, 50, 100, 100)  # 剪切的 左、上、右、下的坐标
region = im.crop(box)
region.show()
```

```python
region = region.transpose(Image.ROTATE_180)  # 剪切部分旋转180度
im.paste(region, box)  # 粘贴到原图box位置
im.show()
```

# 图像滤波

```python
from PIL import ImageFilter
```

```python
im = Image.open('demo.jpg')  # 原图
im
```

```python
im.filter(ImageFilter.DETAIL)  # 精细化
```

```python
im.filter(ImageFilter.BLUR)  # 均值滤波
```

```python
im.filter(ImageFilter.CONTOUR)  # 轮廓图
```

```python
im.filter(ImageFilter.EDGE_ENHANCE) # 边缘锐化
```

```python
im.filter(ImageFilter.EDGE_ENHANCE_MORE)  # 进一步锐化
```

```python
im.filter(ImageFilter.EMBOSS)  # 浮雕处理
```

```python
im.filter(ImageFilter.FIND_EDGES)  # 边缘检查
```

```python
im.filter(ImageFilter.SMOOTH)  # 平滑
```

```python
im.filter(ImageFilter.SMOOTH_MORE)  # 平滑
```

```python
im.filter(ImageFilter.SHARPEN)  # 锐化
```

```python
im.filter(ImageFilter.GaussianBlur)  # 高斯均值
```

```python
im.filter(ImageFilter.MaxFilter)
```

```python
im.filter(ImageFilter.MinFilter)
```
